### To get the players' statistics for a map

```python
url = "https://www.vlr.gg" + "/189055/loud-vs-nrg-esports-champions-tour-2023-americas-league-gf"
resp = requests.get(url, headers=headers)
html = HTMLParser(resp.text)
def get_stats_players(map_info):
    dict_teams_stats = {}
    for table in map_info.css("table.wf-table-inset"):
        for players in table.css("tbody"):
            for player in players.css("tr"):
                player_info = (player.css_first("a").text().replace("\t", "").strip().split())
                player_info.append(player.css_first("a").attributes['href'])
                if player_info[1] not in dict_teams_stats:
                    dict_teams_stats[player_info[1]] = {player_info[0]: {'url': player_info[2]}}
                else:
                    dict_teams_stats[player_info[1]][player_info[0]] = {'url': player_info[2]}

                agents = ""
                for agent in player.css("img"):
                    agents += agent.attributes['title']
                    agents += '/'

                agents = agents[:-1]
                dict_teams_stats[player_info[1]][player_info[0]]['Agents'] = agents

            players_stats = players.text().replace("\t", "").strip().split()
            for i in range(int(len(players_stats)/40)):
                player_name = players_stats[0+(40*i)]
                team_name = players_stats[1+(40*i)]

                dict_teams_stats[team_name][player_name]['Rating'] = players_stats[2+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking Rating'] = players_stats[3+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending Rating'] = players_stats[4+(40*i)]

                dict_teams_stats[team_name][player_name]['ACS'] = players_stats[5+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking ACS'] = players_stats[6+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending ACS'] = players_stats[7+(40*i)]

                dict_teams_stats[team_name][player_name]['K'] = players_stats[8+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking K'] = players_stats[9+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending K'] = players_stats[10+(40*i)]

                dict_teams_stats[team_name][player_name]['D'] = players_stats[12+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking D'] = players_stats[13+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending D'] = players_stats[14+(40*i)]

                dict_teams_stats[team_name][player_name]['A'] = players_stats[16+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking A'] = players_stats[17+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending A'] = players_stats[18+(40*i)]

                dict_teams_stats[team_name][player_name]['+/-'] = players_stats[19+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking +/-'] = players_stats[20+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending +/-'] = players_stats[21+(40*i)]

                dict_teams_stats[team_name][player_name]['KAST'] = players_stats[22+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking KAST'] = players_stats[23+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending KAST'] = players_stats[24+(40*i)]

                dict_teams_stats[team_name][player_name]['ADR'] = players_stats[25+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking ADR'] = players_stats[26+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending ADR'] = players_stats[27+(40*i)]

                dict_teams_stats[team_name][player_name]['HS'] = players_stats[28+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking HS'] = players_stats[29+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending HS'] = players_stats[30+(40*i)]

                dict_teams_stats[team_name][player_name]['FK'] = players_stats[31+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking FK'] = players_stats[32+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending FK'] = players_stats[33+(40*i)]

                dict_teams_stats[team_name][player_name]['FD'] = players_stats[34+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking FD'] = players_stats[35+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending FD'] = players_stats[36+(40*i)]

                dict_teams_stats[team_name][player_name]['FK_FD_+/-'] = players_stats[37+(40*i)]
                dict_teams_stats[team_name][player_name]['Attacking FK_FD_+/-'] = players_stats[38+(40*i)]
                dict_teams_stats[team_name][player_name]['Defending FK_FD_+/-'] = players_stats[39+(40*i)]

    return dict_teams_stats

for map in html.css("div.vm-stats-game"):
    print(get_stats_players(map))
```

```python
def get_aliases(url_path):
    url = "https://www.vlr.gg" + url_path
    resp = requests.get(url, headers=headers)
    html = HTMLParser(resp.text)

    name = html.css_first("span.m-item-team-name").text().strip()
    tag = html.css_first("span.m-item-team-tag").text().strip()

    return name, tag

get_aliases('/team/11195/00nation')
```

```python
def check_if_zero(stats):
    if len(stats) == 0:
        return '0'
    else:
        return stats[0]

def get_performance_stats(url_path):
    url = "https://www.vlr.gg" + url_path
    resp = requests.get(url, headers=headers)
    html = HTMLParser(resp.text)

    dict_stats = {}
    for map in html.css("div.vm-stats-game"):
        map_id = map.attributes['data-game-id']

        try:
            table  = map.css('table.wf-table-inset')[3]
            dict_map = {}
            for row in table.css('tr'):
                if row.text().strip().split()[0] == '2K':
                    header = row.text().strip().split()
                else:
                    data = row.css('td')

                    player_name = data[0].text().strip().split()[0]
                    team_name = data[0].text().strip().split()[1]

                    if team_name not in list(dict_map.keys()):
                        dict_map[team_name] = {}
                    dict_map[team_name][player_name] = {}

                    for i in range(2,len(data)):
                        stats = check_if_zero(data[i].text().strip().split())
                        dict_map[team_name][player_name][header[i-2]] = stats

            dict_stats[map_id] = dict_map
        except:
            break
    return dict_stats

get_performance_stats('/189055/loud-vs-nrg-esports-champions-tour-2023-americas-league-gf/?game=all&tab=performance')
```

### To get the maps' statistics from a match

```python
def update_dicts(general_stats, performance_stats):
    for map in performance_stats:
        for map_2 in general_stats:
            if general_stats[map_2]['Map ID'] == map:
                for team in performance_stats[map]:
                    for team_2 in general_stats[map_2]:
                        if team == team_2 or ((team in team_abbrevs) and (team_2 in team_abbrevs[team])):
                            for player in performance_stats[map][team]:
                                for player_2 in general_stats[map_2][team_2]['Player Stats']:
                                    if player == player_2:
                                        general_stats[map_2][team_2]['Player Stats'][player_2].update(performance_stats[map][team][player])
                                        break
                            break
                break

def get_stats_match(url_path):
    url = "https://www.vlr.gg" + url_path
    resp = requests.get(url, headers=headers)
    html = HTMLParser(resp.text)

    global team_abbrevs

    try:
        dict_match = {}
        teams_names = []
        teams_url = {}
        for team in html.css("a.match-header-link"):
            url_path = team.attributes['href']
            team_name, team_alias = get_aliases(url_path)

            if team_name != team_alias:
                if team_alias not in list(team_abbrevs.keys()):
                    team_abbrevs[team_alias] = [team_name]
                elif team_name not in team_abbrevs[team_alias]:
                    team_abbrevs[team_alias].append(team_name)
            teams_names.append(team_name)
            teams_url[team_name] = url_path

        for map in html.css("div.vm-stats-game"):
            dict_map = {}

            map_id = map.attributes['data-game-id']
            info_map = map.css_first("div.map")
            if info_map != None:
                info_map = info_map.text().replace("\t", "").strip().split("\n")
                map_name = info_map[0]
                dict_map['Duration'] = info_map[-1]

                team_start = [None, None]
                if 'mod-win' in map.css("div.rnd-sq")[0].attributes['class']:
                    if 'mod-t' in map.css("div.rnd-sq")[0].attributes['class']:
                        team_start[0] = 'T'
                        team_start[1] = 'CT'
                    else:
                        team_start[0] = 'CT'
                        team_start[1] = 'T'
                else:
                    if 'mod-t' in map.css("div.rnd-sq")[1].attributes['class']:
                        team_start[1] = 'T'
                        team_start[0] = 'CT'
                    else:
                        team_start[1] = 'CT'
                        team_start[0] = 'T'

                cont = 0
                for info_team in map.css("div.team"):
                    if info_team.css_first('div.team-name') == None:
                        break
                    name_team = info_team.css_first('div.team-name').text().replace("\t", "").strip()
                    score_team = (info_team.css_first('div.score').text().strip())
                    att_score_team = (info_team.css_first('span.mod-t').text())
                    def_score_team = (info_team.css_first('span.mod-ct').text())

                    dict_team = {'Score': score_team, 'Attacking Score': att_score_team, 'Defending Score': def_score_team, 'Start Side': team_start[cont]}
                    cont += 1
                    dict_map[name_team] = dict_team

                dict_players = get_stats_players(map)
                teams = list(dict_players.keys())
                for i in range(len(teams)):
                    if teams[i] not in list(dict_map.keys()):
                        #if teams[i] not in list(team_abbrevs.keys()):
                        #    print(dict_map.keys(), teams[i])

                        for real_name in list(dict_map.keys()):
                            try:
                                name_index = team_abbrevs[teams[i]].index(real_name)
                            except:
                                continue

                        dict_map[team_abbrevs[teams[i]][name_index]]['Player Stats'] = dict_players[teams[i]]
                        dict_map[team_abbrevs[teams[i]][name_index]]['URL'] = teams_url[team_abbrevs[teams[i]][name_index]]
                    else:
                        dict_map[teams[i]]['Player Stats'] = dict_players[teams[i]]
                        dict_map[teams[i]]['URL'] = teams_url[teams[i]]

            else:
                map_name = 'All Maps'

                dict_players = get_stats_players(map)
                teams = list(dict_players.keys())

                for i in range(len(teams)):
                    if teams[i] in list(team_abbrevs.keys()):
                        for real_name in teams_names:
                            try:
                                name_index = team_abbrevs[teams[i]].index(real_name)
                            except:
                                continue

                        dict_map[team_abbrevs[teams[i]][name_index]] = {}
                        dict_map[team_abbrevs[teams[i]][name_index]]['Player Stats'] = dict_players[teams[i]]
                        dict_map[team_abbrevs[teams[i]][name_index]]['URL'] = teams_url[team_abbrevs[teams[i]][name_index]]
                    else:
                        dict_map[teams[i]] = {}
                        dict_map[teams[i]]['Player Stats'] = dict_players[teams[i]]
                        dict_map[teams[i]]['URL'] = teams_url[teams[i]]

            dict_map['Map ID'] = map_id
            dict_match[map_name] = dict_map

        update_dicts(dict_match, get_performance_stats(html.css('a.vm-stats-tabnav-item')[1].attributes['href']))

        return dict_match

    except:
        return dict_match

team_abbrevs = {}
get_stats_match("/189055/loud-vs-nrg-esports-champions-tour-2023-americas-league-gf")
```

### To get the matches' statistics from a number of pages

```python
def get_stats(num_pages, first_page=0):
    matches = {}
    for p in range(first_page,num_pages+first_page):
        url = "https://www.vlr.gg/matches/results/?page=" + str(p+1)
        resp = requests.get(url, headers=headers)
        html = HTMLParser(resp.text)

        for item in html.css("a.wf-module-item"):
            camp = item.css_first("div.match-item-event").text().replace("\t", "").strip().split("\n")[1]
            if ("Challengers" in camp) or ("Champions" in camp):
                url_path = item.attributes['href']
                dict_match = get_stats_match(url_path)
                if dict_match != None:
                    dict_match['Tournament'] = camp
                    dict_match['Score'] = {}
                    for team in item.css("div.match-item-vs-team"):
                        score_info = team.text().replace("\t", "").strip().split("\n")
                        dict_match['Score'][score_info[0]] = score_info[-1]

                    matches[url_path] = dict_match

    return matches


team_abbrevs = {}
matches_data = get_stats(2, 0)
matches_data
```

## Creating DataFrames from the obtained data

The data was obtained calling the *get_stats* funcion in the previous cell and stored in the **matches_data** dictionary.

```python
import pandas as pd

match_info_df = pd.DataFrame(columns=['Match_URL', 'Tournament', 'Team_1', 'Score_1', 'Team_2', 'Score_2'])

for match_url in list(matches_data.keys()):
    camp = matches_data[match_url]['Tournament']
    team_1 = list(matches_data[match_url]['Score'].keys())[0]
    score_1 = matches_data[match_url]['Score'][team_1]
    team_2 = list(matches_data[match_url]['Score'].keys())[1]
    score_2 = matches_data[match_url]['Score'][team_2]
    match_info_df.loc[len(match_info_df)] = [match_url, camp, team_1, score_1, team_2, score_2]

match_info_df
```

```python
map_info_df = pd.DataFrame(columns=['Match_URL', 'Map_Name', 'Map_ID', 'Duration', 'Team_1', 'URL_1', 'Start_Side_1', 'Score_1', 'Attacking_Score_1', 'Defending_Score_1', 'Team_2', 'URL_2', 'Start_Side_2', 'Score_2', 'Attacking_Score_2', 'Defending_Score_2'])

for match_url in list(matches_data.keys()):
    for map in list(matches_data[match_url].keys()):
        if map != 'Tournament' and map != 'Score':
            if map != 'All Maps':
                duration = matches_data[match_url][map]['Duration']

                team_1 = list(matches_data[match_url][map].keys())[1]
                url_1 = matches_data[match_url][map][team_1]['URL']
                side_1 = matches_data[match_url][map][team_1]['Start Side']
                score_1 = matches_data[match_url][map][team_1]['Score']
                att_score_1 = matches_data[match_url][map][team_1]['Attacking Score']
                def_score_1 = matches_data[match_url][map][team_1]['Defending Score']

                team_2 = list(matches_data[match_url][map].keys())[2]
                url_2 = matches_data[match_url][map][team_2]['URL']
                side_2 = matches_data[match_url][map][team_2]['Start Side']
                score_2 = matches_data[match_url][map][team_2]['Score']
                att_score_2 = matches_data[match_url][map][team_2]['Attacking Score']
                def_score_2 = matches_data[match_url][map][team_2]['Defending Score']

                map_id = matches_data[match_url][map]['Map ID']
                map_info_df.loc[len(map_info_df)] = [match_url, map, map_id, duration, team_1, url_1, side_1, score_1, att_score_1, def_score_1, team_2, url_2, side_2, score_2, att_score_2, def_score_2]

map_info_df
```

```python
player_info_df = pd.DataFrame(columns=['Match_URL', 'Map_ID', 'Team_URL', 'Player_Name', 'Player_URL', 'Agents', 'Rating', 'Attacking Rating', 'Defending Rating', 'ACS', 'Attacking ACS', 'Defending ACS', 'K', 'Attacking K', 'Defending K', 'D', 'Attacking D', 'Defending D', 'A', 'Attacking A', 'Defending A', '+/-', 'Attacking +/-', 'Defending +/-', 'KAST', 'Attacking KAST', 'Defending KAST', 'ADR', 'Attacking ADR', 'Defending ADR', 'HS', 'Attacking HS', 'Defending HS', 'FK', 'Attacking FK', 'Defending FK', 'FD', 'Attacking FD', 'Defending FD', 'FK_FD_+/-', 'Attacking FK_FD_+/-', 'Defending FK_FD_+/-', '2K', '3K', '4K', '5K', '1v1', '1v2', '1v3', '1v4', '1v5', 'ECON', 'PL', 'DE'])

for match_url in list(matches_data.keys()):
    for map in list(matches_data[match_url].keys()):
        if map != 'Tournament' and map != 'Score':
            map_id = matches_data[match_url][map]['Map ID']
            for team in list(matches_data[match_url][map].keys()):
                if team != 'Map ID' and team != 'Duration':
                    team_url = matches_data[match_url][map][team]['URL']
                    for player in list(matches_data[match_url][map][team]['Player Stats'].keys()):
                        data = [match_url, map_id, team_url, player]
                        for stats in list(matches_data[match_url][map][team]['Player Stats'][player].keys()):
                            data.append(matches_data[match_url][map][team]['Player Stats'][player][stats])
                        try:
                            player_info_df.loc[len(player_info_df)] = data
                        except:
                            continue

player_info_df
```
